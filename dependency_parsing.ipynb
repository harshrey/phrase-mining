```python
# convert a text file into a processed Doc object.
file_name = 'introduction.txt'
file_text = open(file_name).read()
file_doc = nlp(file_text)
print ([token.text for token in file_doc])
```

```python
#identify and separate sentences using a delimiter, here fullstop. 

sentences = list(file_doc.sents)
print(len(sentences))

for sentence in sentences:
    print (sentence)
```

###### Tokenisation

```python
for token in file_doc:
    print (token, token.idx)
```

###### Remove Punctuations

```python
punct_doc = [token for token in file_doc if not token.is_punct]
>>> print (punct_doc)
```

###### Remove Stopwords

```python
stopword_doc = [token for token in punct_doc if not token.is_stop]
>>> print (stopword_doc)
```

###### Lemmatization

```python
def preprocess_token(token):
    # Reduce token to its lowercase lemma form and make it lower case. 
    return token.lemma_.strip().lower()


filtered_doc = [preprocess_token(token) for token in stopword_doc]
print(filtered_doc)
```

###### Word Frequency

```python
from collections import Counter

word_freq = Counter(filtered_doc)

# 5 commonly occurring words with their frequencies-to understand what the text is about
common_words = word_freq.most_common(5)
print (common_words)

# Unique words
unique_words = [word for (word, freq) in word_freq.items() if freq == 1]
print (unique_words)
```

```python
#without punctuations :
words_all = [token.text for token in stopword_doc if not token.is_punct]
word_freq_all = Counter(words_all)
# 5 commonly occurring words with their frequencies
common_words_all = word_freq_all.most_common(5)
print (common_words_all)
```

```python
#Part of Speech tagging
for token in stopword_doc:
     print (token, token.tag_, token.pos_, spacy.explain(token.tag_))
```

```python
#Identifying nouns and adjectives

nouns = []
adjectives = []
for token in stopword_doc:
    if token.pos_ == 'NOUN':
         nouns.append(token)
    if token.pos_ == 'ADJ':
         adjectives.append(token)

print("Nouns: " ,nouns)

print("Adjectives:" ,adjectives)
```

###### Rule-based matching
 Its used to identify and extract tokens and phrases according to patterns and grammatical features (such as part of speech).

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
def extract_full_name(file_doc):
    pattern = [{'POS': 'PROPN'}, {'POS': 'PROPN'}]  #extract a name with this pattern in the data
    matcher.add('FULL_NAME', None, pattern)
    matches = matcher(file_doc)
    for match_id, start, end in matches:
        span = file_doc[start:end]
        return span.text

extract_full_name(file_doc)
```

###### Dependency parsing
Extracting the dependency parse of a sentence to represent its grammatical structure. It defines the dependency relationship between headwords and their dependents. The head of a sentence has no dependency and is called the root of the sentence. 

```python
from spacy import displacy
text = 'I am playing badminton'
doc = nlp(text)
for token in doc:
    print (token.text, token.tag_, token.head.text, token.dep_)
displacy.serve(doc, style='dep')
```

```python
#for entire text!
for token in file_doc:
     print (token.text, token.tag_, token.head.text, token.dep_)
```

##### Traversing a dependency parse tree

```python

#Example: For the word 'leading' in the text

# Extract children of `leading`
print([token.text for token in file_doc[16].children])

# Extract previous neighboring node of `leading`
print (file_doc[16].nbor(-1))

# Extract next neighboring node of `leading`
print (file_doc[16].nbor())

# Extract all tokens on the left of `leading`
print([token.text for token in file_doc[16].lefts])

# Extract tokens on the right of `leading`
print([token.text for token in file_doc[16].rights])

# Print subtree of `leading`
print (list(file_doc[16].subtree))
```

##### Shallow parsing, or chunking, is the process of extracting phrases from unstructured text. Chunking groups adjacent tokens into phrases on the basis of their POS tags. 

```python
import textacy
text = ('The code will introduce reader about use of dependency grammar')
pattern = r'(<VERB>?<ADV>*<VERB>+)' #we can use any combination
doc = textacy.make_spacy_doc(text, lang='en_core_web_sm')
verb_phrases = textacy.extract.pos_regex_matches(doc, pattern)
# Print all Verb Phrase, what is the sentence about
for chunk in verb_phrases:
     print(chunk.text)


# Extract Noun Phrase to explain what nouns are involved
for chunk in doc.noun_chunks:
    print (chunk)
```
